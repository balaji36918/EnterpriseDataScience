```python
import pandas as pd 
import numpy as np

%matplotlib inline

import matplotlib as mpl 
import matplotlib.pyplot as plt

import seaborn as sns

import plotly
import plotly.graph_objects as go 
```

```python
country_list = ['Germany', 'India', 'United States']
# Importing the data for the world population

wp = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
wp.head()
```

```python
#wp = wp.rename(columns = {'Country_Region':'Country'}
#Converting the date to string using 'datetime'.
wp['date'] = pd.to_datetime(wp['date'])
pivot = pd.pivot_table(data = wp,
                                  index = 'date',
                                  columns = 'location',
                                  values = 'total_cases',)

piot = pivot.fillna(0)
```

```python
pivot= pivot.replace(np.nan,0)

pivot.reset_index(inplace = True)

pivot.head()
```

```python
time_idx= pivot["date"]
```

```python
df_plot = pd.DataFrame({'Date': time_idx})
df_plot.head()
```

```python
country_list = ['Germany', 'India', 'United States']
# for each in country_list:
#     df_plot[each]=np.array(wp[wp['location']==each]iloc[:].sum(axis=0))

for each in country_list:
    df_plot[each] = pivot[each]
    var = np.array(wp.loc[wp["location"]])
    df_plot[each] = df_plot[each]/var[0]*100 
    
df_plot.head()
```
